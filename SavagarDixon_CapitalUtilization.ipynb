# Notebook Overview

The logic of the notebook is as follows (Chapter 13 of [De La Fuente's](https://books.google.co.uk/books?id=YyW6V_bNHh4C&lpg=PP1&dq=inauthor%3A%22Angel%20de%20la%20Fuente%22&pg=PA643#v=onepage&q=mathematica&f=false) textbook has a *Mathematica* exercise that follows a similar approach) 
1. Set the values of exogenous parameters including two values of technology parameter A.
2. Calculate the steady state values of endogenous parameters $\{C,e,K,n\}$ given the exogenous specification.
  * We will calculate two different steady states for a low and a high value of technology $A$. We are interested in the solution of the model with high technology, but we want to start the economy off from a position of worse productivity.
3. Solve the model for timepaths of $\{C,e,K,n \}$
  1. Symbolically specify the right-hand side of the dynamical system $\{\dot{C},\dot{e},\dot{K},\dot{n} \}$ and take its derivative with respect to $\{C,e,K,n \}$. This gives the Jacobian matrix.
  2. Substitute in exogenous parameters for the specification we are interested in, also substitute in the steady state values of endogeous parameters for this exogenous parameter specification. This gives the Jacobian matrix evaluated at steady state. We substitute in the high productivity $A$ and corresponding steady state as we are interested into transition to this new technology.
4. Given the solution of the model, we know how $\{C,e,K,n \}$ behave over $t$ for a given initial position. We give the inital position as the steady state of the model with a worse technology. Therefore our time paths show what happens if the economy is at rest with a given worse technology, but then converges to a new steady state that arises with better technology.
5. Plot how productivity moves to steady state from this initial position relative to it's long-run steady-state level. 

# Model with Capital Utilization and Capacity Utilization
\begin{align}
\dot{C}(t) &= C(r_t - \rho) = C_t(R_t-\delta_t-\rho) \\
\dot{e}(t) &= r_t e_t - \frac{\pi_t}{\gamma} \\
\dot{K}(t) &= - \delta_t K_t + y_t - \frac{\gamma e^2}{2} - C_t, \quad \text{where} \; I = \dot{K} + \delta_tK_t + \frac{\gamma e_t^2}{\gamma} \\
\dot{n}(t) &= e_t \\
R_t &= (1-\zeta) AF_k(\bar{u}k,l) \\
w_t &= (1-\zeta) AF_l(\bar{u}k,l)  
\end{align}
 
 
Substituting in functional forms
\begin{align}
\dot{C}(t) &= C_t \left( (1-\zeta)A \alpha u_t^\alpha k_t^{\alpha-1} l_t^\beta - \left( \frac{A \alpha k_t^{\alpha-1} l_t^\beta}{\vartheta z^{\frac{\alpha}{\vartheta}}} \right)^{\frac{\vartheta}{\vartheta-\alpha}}-\rho \right) \\
\dot{e}(t) &= r_te_t- \frac{A(u_tk_t)^\alpha l_t^\beta (1-\nu(1-\zeta))-\phi}{\gamma} \\
\dot{K}(t) &= - \left( \frac{A \alpha l_t^\beta k_t^{\frac{\alpha (\vartheta -1)}{\vartheta}}}{\vartheta z^{\frac{\alpha}{\vartheta}}} \right)^{\frac{\vartheta}{\vartheta-\alpha}} + A (u_tk_t)^\alpha l_t^\beta - \phi - \frac{\gamma e_t^2}{2} - C_t \\
\dot{n}(t) &= e_t \\
R(t) & = (1-\zeta) A u_t^\alpha \alpha k_t^{\alpha-1} l_t^\beta = r_t + \delta_t\\
w(t) &= (1- \zeta) A (u_tk_t)^\alpha \beta l_t^{\beta -1}\\
\pi_t &= A(u_tk_t)^\alpha l_t^\beta (1-\nu(1-\zeta))-\phi \\
\delta_t &= \left( \frac{A \alpha k_t^{\alpha-1} l_t^\beta}{\vartheta z^{\frac{\alpha}{\vartheta}}} \right)^{\frac{\vartheta}{\vartheta-\alpha}}\\
r_t &=  A \alpha u_t^\alpha k_t^{\alpha-1} l_t^\beta - \left( \frac{A \alpha k_t^{\alpha-1} l_t^\beta}{\vartheta z^{\frac{\alpha}{\vartheta}}} \right)^{\frac{\vartheta}{\vartheta-\alpha}}\\
u_t &= \left( \frac{A \alpha k_t^{\alpha-1}l_t^\beta}{\vartheta z} \right)^{\frac{1}{\vartheta-\alpha}}
 \end{align}

```python
K, n, C, e, L, alpha, beta, phi, gamma, A, sigma, zeta, rho, xi, eta, z, vartheta, u = symbols("K n C e L alpha beta phi gamma A sigma zeta rho xi eta z vartheta u")
```

```python
#Make a list of the parameterization
alpha1 = 0.3
beta1  = 0.5
phi1   = 0.3 #0.3
gamma1 = 50.0 # 3.0 
sigma1 = 1.0 #1.5 #changed
rho1   = 0.025
#xi1    = 0.01
eta1   = 0.5 #0.5
zeta1  = 0.2 # Correponds to markup mu of 1.05 and zeta1=0.09 correponds to mu=1.1. Put to 0.4 for a reasonably
#ponounced effect.
z1     = 1.0 
vartheta1 = 1.4 


#Shock to A
A1     = 1.01 #we shall solve the model for this technology but ask what happens if we begin in SS that arises
# under a worse technology.

# To get xi such that L^* = 1
nu1   = alpha1 + beta1
y_ss1 = (nu1 * (1 - zeta1) * phi1)/(1 - (1-zeta1)*nu1)
c_ss1 = y_ss1*(1- (alpha1/vartheta1)) - (alpha1*phi1)/vartheta1
k_ss1 = (alpha1 * (y_ss1 + phi1) * (1 - zeta1 - (1/vartheta1) ))/(rho1)
u_ss1 = ((alpha1*(y_ss1 + phi1))/(vartheta1 * z1 * k_ss1))**(1/vartheta1)
l_ss1 = ((y_ss1+phi1)/(A1 * (u_ss1*k_ss1)**alpha1))**(1/beta1)


xi1    =  ((((1-zeta1)*beta1*(y_ss1+phi1))/(l_ss1**(1+eta1)))**(1/sigma1)*(1/c_ss1))**sigma1 \
* ((y_ss1+phi1)/(A1 * (u_ss1*k_ss1)**alpha1))**((eta1+sigma1)/beta1)

Astart     = 1.0


params = [(zeta, zeta1), 
          (alpha, alpha1),
          (beta, beta1),
          (phi, phi1),
          (gamma, gamma1),
          (A, A1),
          (sigma, sigma1),
          (rho, rho1),
          (xi, xi1),
          (eta, eta1),
          (z, z1),
          (vartheta, vartheta1)] #vartheta should exceed 1
```

## Steady State

From the $\dot{e}$ equation which gives $\pi=0$ we get

\begin{align}
y^* = \frac{\nu  (1 - \zeta)  \phi}{1 - (1-\zeta)\nu}
\end{align}

From the $\dot{C}$ equation which gives $r=\rho$ we get

\begin{align}
  k^* = \frac{\alpha (y^* + \phi)}{\rho} \left(1 - \zeta - \frac{1}{\vartheta} \right)
\end{align}

helps us to get utilization

\begin{align}
  u^* = \left( \frac{\alpha (y^*+\phi)}{\vartheta z k^*} \right)^{\frac{1}{\vartheta}}
\end{align}
so combined we can get utilized capital $u^*k^*$.

By rearranging $\dot{e}=0$ which is the $\pi=0$ condition then labour per firm looks similar to before except now we have $uk$ not simply $k$
\begin{align}
  l^* = \left( \frac{y^* + \phi}{ A(u^*k^*)^\alpha} \right)^{\frac{1}{\beta}}
\end{align}

Finally the $\dot{K}=0$ equation gives $c^*=\frac{C^*}{n^*}= y^*- \delta^* k^*$

\begin{align}
  c^* = y^* \left(1 - \frac{\alpha}{\vartheta} \right) - \frac{\alpha \phi}{\vartheta}
\end{align}
We can see that compared to the original case $k^*, c^*$ are strictly less and $l^*$ is more.

Finally the key to solution is the intratemporal condition which looks exactly the same as the previous case except of course $c^*$ and $l^*$ are different.

\begin{align}
  n^* = \left[ \left( \frac{(1-\zeta) \beta (y^*+\phi)}{\xi l^{* 1 + \eta}} \right)^{\frac{1}{\sigma}}\frac{1}{c^*} \right]^{\frac{\sigma}{\eta+\sigma}}
\end{align}

Then with $n^*$ we can gain all the aggregate variables as $K^* =
k^*n^*$ and $L^* = l^*n^*$ and $C^* = c^*n^*$ and we know $e^* = 0$
which is immediate from $\dot{n}=0$

(To get $\xi$ in the calibration above note $L^*= n^* l^*$, and since only $n^*$ is a function of $\xi$, set $L^*=1$ then multiply both sides by $\xi^{\frac{1}{\eta+\sigma}}$ to get $\xi^{\frac{1}{\eta+\sigma}} = \xi^{\frac{1}{\eta+\sigma}}n^* l^* = \left[ \left( \frac{(1-\zeta) \beta (y^*+\phi)}{l^{* 1 + \eta}} \right)^{\frac{1}{\sigma}}\frac{1}{c^*} \right]^{\frac{\sigma}{\eta+\sigma}} l^*$ where the right-hand side is purged of any dependence on $\xi$.)

```python
def steady_state(alpha, beta, phi, gamma, A, sigma, rho, xi, eta, zeta, z, vartheta):
    """
    Calculate the steady state for a given set of parameters.
    """
    nu   = alpha + beta

    y_ss = (nu * (1 - zeta) * phi)/(1 - (1-zeta)*nu)

    k_ss = (alpha * (y_ss + phi) * (1 - zeta - (1/vartheta) ))/(rho)

    u_ss = ((alpha*(y_ss + phi))/(vartheta * z * k_ss))**(1/vartheta)

    l_ss = ((y_ss+phi)/(A * (u_ss*k_ss)**alpha))**(1/beta)

    c_ss = y_ss*(1- (alpha/vartheta)) - (alpha*phi)/vartheta

    n_ss = ((((1-zeta)*beta*(y_ss+phi))/(xi*l_ss**(1+eta)))**(1/sigma)*(1/c_ss))**(sigma/(eta+sigma))

    K_ss = k_ss * n_ss

    L_ss = l_ss * n_ss

    C_ss = c_ss * n_ss

    e_ss = 0.0
    
    delta_ss = z * u_ss**vartheta

    return matrix([[C_ss], [e_ss], [K_ss], [n_ss], [L_ss], [u_ss], [delta_ss]])
```

```python
SS_A1 = steady_state(alpha1, beta1, phi1, gamma1, A1, sigma1, rho1, xi1, eta1, zeta1, z1, vartheta1);
SS_A1
```

# Dynamics
In this section we solve for the dynamics around the steady state corresponding to technology `A1`.

Our experiments begin the model from an initial position that corresponds to steady state for a different value of techonology `Astart`. We then study transition dynamics.

```python
L = (((1-zeta) * A**(vartheta/(vartheta-alpha)) * (alpha/(vartheta*z))**(alpha/(vartheta-alpha)) * K**(alpha*(vartheta-1)/(vartheta-alpha)) * beta * n**((vartheta*(1-(alpha+beta)))/(vartheta-alpha)))/(xi * C**sigma))**((vartheta-alpha)/(vartheta-alpha-beta*alpha*(1+eta-beta))) 

u = ((A * alpha * K**(alpha-1) * L**(beta) * n**(1-(alpha + beta)))/(vartheta * z ))**(1/(vartheta-alpha))

delta = z * u**vartheta


# delta = 0.0

R = (1 - zeta) * A * alpha * u**alpha * K**(alpha - 1) * L**beta * n**(1 - (alpha +beta))

r = R - delta

pi = A * (u * K)**alpha * L**beta * n**-(alpha + beta) * (1 - (1 - zeta) * (alpha + beta)) - phi

# r = rho
Cdot = C / sigma * ( r - rho)

#pi = 0
edot = r * e - 1 / gamma * pi

#Y = C
Kdot = -delta * K + n * (A * (u*K)**alpha * L**beta * n**-(alpha + beta) - phi) - 0.5 * gamma * e**2 - C  

#e = 0
ndot = e
```

```python
Z = Matrix([Cdot, edot, Kdot, ndot]);
Q = Matrix([C, e, K, n])
jacb = Z.jacobian(Q); #Jacobian is a SymPy function
```

```python
#Z_num = Matrix([Cdot_num, edot_num, Kdot_num, ndot_num]);
jacb_num = jacb.subs(params)
jacb_num_ss = jacb_num.subs([(C, SS_A1[0]), (e, SS_A1[1]), (K, SS_A1[2]), (n, SS_A1[3])])
```

```python
evals, evecs = LA.eig(matrix2numpy(jacb_num_ss, dtype=float)); evals
```

```python
evals_stable = evals[2:4]; 
evals_stable = matrix(evals_stable);
```

```python
#such that the column v[:,i] is the eigenvector corresponding to the eigenvalue w[i].
evecs_stable = evecs[:,2:4]; 
evecs_stable = matrix(evecs_stable); 
```

```python
def eigval_exp_diag(t):

    return diag([exp(evals[2]*t),exp(evals[3]*t)], k=0)  
```

## Start Position

We begin the model from a position in the state space that corresponds to the the steady state for a different technology parameter. Then we observe how the dynamics of our solved model pull it towards steady state.

The *modified Solow Residual* (SR) measure is equivalent to the plain no capital utilization model in steady state. This is because the modified measure adjusts for the capital utilization process. The relationship between uncorrected (uc) measure of productivity and corrected (c), or modified SR,  is 
\begin{align}
\mathcal{P}^{uc} = \frac{y^*}{(k^{*\alpha} l^\beta)^{\frac{1}{\alpha+\beta}}}= \frac{y^*}{((\mathfrak{u}^*k^*)^\alpha l^\beta)^{\frac{1}{\nu}}}\mathfrak{u}^{\frac{\alpha}{\nu}} = \mathcal{P}^{c}\mathfrak{u}^{\frac{\alpha}{\nu}}  
\end{align}


```python
SS_Astart = steady_state(alpha1, beta1, phi1, gamma1, Astart, sigma1, rho1, xi1, eta1, zeta1, z1, vartheta1); SS_Astart

Prod_SS_Astart_corrected   = Astart**(1/(alpha1+beta1)) * (alpha1+beta1) * (1-zeta1) * ((1-(alpha1+beta1)*(1-zeta1))/phi1)**((1-alpha1-beta1)/(alpha1+beta1))

Prod_SS_Astart_uncorrected = Prod_SS_Astart_corrected * SS_Astart[5,0]**(alpha1/(alpha1+beta1))

[SS_Astart, Prod_SS_Astart_corrected, Prod_SS_Astart_uncorrected]
```

```python
SS_dev = SS_Astart - SS_A1; SS_dev #L, u and delta do not change in response to the change in
#technology in long run
```

```python
def solution_traj(t):
    
    return SS_A1[0:4] + evecs_stable * eigval_exp_diag(t) * LA.inv(evecs_stable[2:4,:]) * SS_dev[2:4]
```

```python
t_sol_length = 1000


C_solution = zeros(t_sol_length);
e_solution = zeros(t_sol_length);
K_solution = zeros(t_sol_length);
n_solution = zeros(t_sol_length);

CeKn_solution = zeros([4,t_sol_length]);


for i in range(t_sol_length):
    CeKn_solution[:,i] = array(solution_traj(i))[:,0]
    
C_solution = CeKn_solution[0,:];
e_solution = CeKn_solution[1,:];
K_solution = CeKn_solution[2,:];
n_solution = CeKn_solution[3,:];
```

```python
SS_Astart_ar = squeeze(asarray(SS_Astart))

t = range(120)

t_before_shock = int(size(t) * 0.1)


C_jump_solution = concatenate(([SS_Astart_ar[0]]*t_before_shock, C_solution), axis=0)
e_jump_solution = concatenate(([SS_Astart_ar[1]]*t_before_shock, e_solution), axis=0)
K_jump_solution = concatenate(([SS_Astart_ar[2]]*t_before_shock, K_solution), axis=0)
n_jump_solution = concatenate(([SS_Astart_ar[3]]*t_before_shock, n_solution), axis=0)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex='col', figsize=(15, 7))


ax1.plot(t, ((C_jump_solution[:size(t)]/C_jump_solution[0])-1.0)*100, 'r')
ax1.set_xlabel('t')
ax1.set_ylabel('Consumption (C)')

ax2.plot(t, e_jump_solution[:size(t)], 'r') #Measures slope of n, not in % dev.
ax2.set_xlabel('t')
ax2.set_ylabel('Entry (e)')

ax3.plot(t, ((K_jump_solution[:size(t)]/K_jump_solution[0])-1.0)*100, 'r')
ax3.set_xlabel('t')
ax3.set_ylabel('Capital (K)')

ax4.plot(t, ((n_jump_solution[:size(t)]/n_jump_solution[0])-1.0)*100, 'r')
ax4.set_xlabel('t')
ax4.set_ylabel('Number of firms (n)')
```

```python
# Remember L solution and TFP solution change because of utilization

# Since both L and u can be written entirely in K, n, C terms it is possible to find their solutions by substitution
L_solution = (((1-zeta1) * A1**(vartheta1/(vartheta1-alpha1)) * (alpha1/(vartheta1*z1))**(alpha1/(vartheta1-alpha1)) \
               * K_solution**(alpha1*(vartheta1-1)/(vartheta1-alpha1)) * beta1 \
               * n_solution**((vartheta1*(1-(alpha1+beta1)))/(vartheta1-alpha1)))/(xi1 \
                * C_solution**sigma1))**((vartheta1-alpha1)/(vartheta1-alpha1-beta1*alpha1*(1+eta1-beta1))) 

u_solution = ((A1 * alpha1 * K_solution**(alpha1-1) * L_solution**(beta1) \
               * n_solution**(1-(alpha1 + beta1)))/(vartheta1 * z1 ))**(1/(vartheta1-alpha1))
```

```python
meas_prod_A1_uncorrected = (A1 * (u_solution * K_solution/n_solution)**alpha1 \
                            * (L_solution/n_solution)**beta1 - phi1)/((K_solution/n_solution)**alpha1 \
                            * (L_solution/n_solution)**beta1)**(1/(alpha1+beta1))

meas_prod_A1_corrected = (A1 * (u_solution * K_solution/n_solution)**alpha1 \
                            * (L_solution/n_solution)**beta1 - phi1)/((u_solution * K_solution/n_solution)**alpha1 \
                            * (L_solution/n_solution)**beta1)**(1/(alpha1+beta1))

Prod_jump_solution_uncorrected = concatenate(([Prod_SS_Astart_uncorrected]*t_before_shock, meas_prod_A1_uncorrected), axis=0)
Prod_jump_solution_corrected = concatenate(([Prod_SS_Astart_corrected]*t_before_shock, meas_prod_A1_corrected), axis=0)
```
